```python
print('–ü–µ—Ä–µ—Å–µ—á–µ–Ω–∏–µ —Å–∞–º—ã—Ö —Å–ª–æ–≤–æ—ë–º–∫–∏—Ö –∏ —Å–∞–º—ã—Ö –¥–ª–∏–Ω–Ω—ã—Ö —Ç–µ–≥–æ–≤ (top-5)')
%timeit tags.most_words_and_longest(5)
intersection = tags.most_words_and_longest(5)
print('–ù–∞–π–¥—ë–º —Ç–µ–≥–∏, –∫–æ—Ç–æ—Ä—ã–µ –æ–¥–Ω–æ–≤—Ä–µ–º–µ–Ω–Ω–æ –∏ —Å–ª–æ–≤–æ—ë–º–∫–∏–µ, –∏ –¥–ª–∏–Ω–Ω—ã–µ. –≠—Ç–æ –º–æ–∂–µ—Ç —É–∫–∞–∑—ã–≤–∞—Ç—å –Ω–∞ –æ—Å–æ–±–µ–Ω–Ω–æ –ø–æ–¥—Ä–æ–±–Ω—ã–µ –æ–ø–∏—Å–∞–Ω–∏—è.')
for tag in intersection:
    print(f"'{tag}'")
```

```python
print('–¢–æ–ø-5 —Å–∞–º—ã—Ö –ø–æ–ø—É–ª—è—Ä–Ω—ã—Ö —Ç–µ–≥–æ–≤')
%timeit tags.most_popular(5)
popular = tags.most_popular(5)
print('–ü–æ—Å–º–æ—Ç—Ä–∏–º, –∫–∞–∫–∏–µ —Ç–µ–≥–∏ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª–∏ —Å—Ç–∞–≤—è—Ç —á–∞—â–µ –≤—Å–µ–≥–æ. –≠—Ç–æ –ø–æ–º–æ–∂–µ—Ç –ø–æ–Ω—è—Ç—å, –∫–∞–∫–∏–µ –∞—Å–ø–µ–∫—Ç—ã —Ñ–∏–ª—å–º–æ–≤ –Ω–∞–∏–±–æ–ª–µ–µ –ø–æ–ø—É–ª—è—Ä–Ω—ã.')
for tag, count in popular.items():
    print(f"'{tag}' ‚Äî {count} —Ä–∞–∑")
```
